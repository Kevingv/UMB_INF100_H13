## Hvorfor Python? ##

- Programmeringsspråk utviklet av [Guido van Rossum](http://en.wikipedia.org/wiki/Guido_van_Rossum) siden 1989

# Programmering og Tegning med Turtle Graphics

## Vår første tegning ##

Marker cellen med koden under og trykk Shift-Enter!

```python
import turtle
tor = turtle.Turtle()
```

### Hva har vi gjort? ###

1. Vi har lastet inn Python-komponenten `turtle`
1. Vi har skapt en skilpadde og gitt Python beskjed om at den skal hete `tor`
1. `tor` står midt på arket og ser mot høyre

NB: Av tekniske grunner åpnes "arket" skilpadden skal løpe på i et annet vindu. Du kan flytte nettleservinduet med Notebooken og vinduet som viser `tor` slik at de ikke overlapper, før du går til neste celle! Vinduet som viser `tor` kan virke dødt, og du kan ikke endre størrelsen på den, men bare gå videre til neste celle! 

```python
tor.forward(100)
```

Vi har gitt `tor` et kommando: han er gått 100 skritt frem og tegnet et strek underveis.

Vi kan også legge flere kommandoer i en celle (trykk altid Shift-Enter for å utføre koden i en celle):

```python
tor.right(60)
tor.forward(200)
tor.left(90)
tor.forward(50)
```

### Hvordan styrer vi skilpadden? ###

1. Vi gir ett kommano (en instruksjon) ad gangen
1. Hver instruksjon har følgende struktur
        <skilpaddens navn>.<instruksjon>(<informasjon>)
   - skilpaddens navn og instruksjonen forbindes med `.` (punktum)
   - informasjonen som følger er avhengig av instruksjonen som gis, f eks
     - `forward(100)`: informasjonen er skrittlengden (i piksel)
     - `right(90)`: informasjonen er vinkelen (i grader)
   - noen kommandoer trenger ingen informasjon, men parentesene må være med
   - informasjonen som gis i parentesene kalles også for *argumenter*

Her et eksempel som viser flere instruksjoner

```python
tor.left(90)
tor.forward(100)
tor.up()         # løft pennen, neste kommando tegner ikke strek
tor.forward(50)
tor.down()       # senk pennen
tor.forward(50)
tor.left(60)
tor.width(5)
tor.color('blue')
tor.forward(90)
tor.color(1.0, 0.3, 0.8)
tor.forward(70)
```

#### Hva har vi gjort? ####

1. `up()` og `down()` løfter og senker pennen; det tegnes ikke noe så lenge pennen er oppe
1. alt som kommer etter en `#` er en *kommentar*: informasjon for programmereren, som Python ignorerer
1. `width()` kommandoen setter strekbredden
1. `color()` kommandoen setter strekfargen; fargen oppgis enten
   - som navn på engelsk (husk hermetegn rund navnet)
   - som rød, grønn og blå (rgb) fargeintensitet (mellom 0 og 1; bruk desimal*punkt*, ikke -komma)
   
### Hvilke kommandoer forstår tor og hvilken informasjon trenger de? ###

- For å se hvilke kommandoer `tor` forstår, gå til en `In` celle, skriv inn (husk punktum!)
        tor.
  og trykk `Tab` (tabulator) tasten. Du får opp en rullegardin med kommandoer
- Velg en kommando fra rullegardinen med Enter
- For å finne ut hva kommandoen gjør, sett et spørsmålstegn bak og trykk Shift-Enter

```python
tor.backward?
```

- I hjelpevinduet som kommer opp viser `Definition` feltet hvordan kommandoen skal brukes
- I hjelpevinduet for `backward()` kommandoen ser du følgende `Definition`
        tor.backward(self, distance)
- **Viktig**: `self` argumentet står der av tekniske grunner, det skal **ikke** oppgis når kommandoen brukes, dvs vi bruker `backward()` som
        tor.backward(50)
- Det samme gjelder alle andre kommandoer: se bort fra `self`!

### Hva skjer hvis jeg gjør en feil? ###

Det finnes to typer feil:

- Tor gjør noe annet enn du ønsker: Prøv å forandre på kommandoen eller et annet kommando!
- Tor gjør ikke noe og du får en feilmelding

```python
tor.color('min yndlingsfarge')
```

- Feilmeldingene kan være lang
- Se på bunnen (det er punktet der ting virkelig skjæret seg) og på toppen (da er vi veldig nær kommandoen du har gitt)
- Her ser vi på bunnen
        TurtleGraphicsError: bad color string: min yndlingsfarge
- Det betyr at Turtle Graphics ikke forstår hva du mener med `'min yndlingsfarge'`: du må oppgi fargenavn på engelsk (og da kan det også være at TG ikke kjenner navnet)!
- Prøv gjerne å google feilmeldingen!

### Hva skjer hvis jeg gir feil antall argumenter? ###

```python
tor.forward()
```

- Python klager på at `forward()` ønsker **to** argumenter, men at vi bare har gitt **ett**
- Men vi har jo gitt **null** argumenter?!?
- **OBS**: Bak kulissene legger Python altid `self` til, så i feilmelding ovenfor må vi trekke fra 1, dvs tenke oss meldingen som "takes exactly 1 argument (0 given)"

## Og hvor er Tina? ##

```python
tina = turtle.Turtle()
tina.color('orange')
tina.width(2)
tina.circle(100)
```

Med `pos()` kommandoen kan vi finne ut hvor `tina` befinner seg

```python
tina.pos()
```

Med `goto()` kan vi sende henne hvor som helst på skjermen

```python
tina.goto(40, 50)
```

Vi kan kombinere begge deler:

```python
x, y = tina.pos()         # få ut posisjon og huske x og y koordinatene
tina.goto(x+100, y-150)   # send tina til et nytt sted, 100 pixler til høyre og 150 nedover
```

## Litt om IPython Notebook

1. Når du starter opp ipython med
        ipython notebook --pylab=inline
   - startes det en ipython server som går bak kulissene 
   - serveren er tilgjengelig som en webserver på lokal maskin
   - en nettleser kobler seg automatisk mot serveren og viser IPython Dashbord
   - dashbordet viser alle IPython Notebooks (`*.ipynb` filer) som ligger i mappen der du started IPython
   - derfor bør du starte IPython fra din `INF100/Python` mappe
1. Arbeid med Notebooks
   - dobbelklikk på en celle for å redigere den
   - Shift-Enter går videre til neste celle (utfører koden i kodeceller)
   - Ctrl-Enter kjører koden og blir stående i samme celle
   - velg celletype (code, markdown (=tekst), heading) i rullegardin på toppen
   - bruk "Insert" for å legge til flere celler
   - se Help > Keyboard Shortcuts for å se hvordan du kan spare mange mus-mil
1. Lagring av Notebooks
   - for å endre navnet, klikk på navnet helt på toppen, ved siden av "IP[y]: Notebook"
   - for å lagre, klikk på diskettsymbolet eller velg File > Save and Checkpoint
   - filene lagres i oppstartmappen
1. Avsluting
   - lagre først
   - velg File > Close and halt
   - lukk dashbord-vinduet i nettlesern
   - trykk Ctrl-C i vinduet der du startet serveren eller lukk det vinduet
1. Mer informasjon
   - se help over
   - [IPython Notebook nettside](http://ipython.org/notebook.html)
